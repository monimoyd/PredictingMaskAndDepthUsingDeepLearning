```python
%matplotlib inline
from matplotlib import pyplot as plt

def plot_metrics(epochs, metric_list, title, x_label, y_label):
    epoch_list = [epoch + 1 for epoch in range(epochs)]
    fig, axs = plt.subplots(figsize=(5,5))
    axs.plot(epoch_list,metric_list)
    axs.set_title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.show()
    
EPOCHS =5
```

```python
epoch_training_time_list =[3600.63835811615,  3622.2514436244965, 3626.0100643634796, 3661.500823497772, 3653.637369155884]
plot_metrics(EPOCHS, epoch_training_time_list, "Epoch vs Training Time", "Epoch", "Training Time")
```

```python
epoch_data_loading_time_list = [10.604703187942505, 10.590641975402832, 10.588401794433594, 10.634032964706421, 10.622773170471191]
plot_metrics(EPOCHS, epoch_data_loading_time_list, "Epoch vs Data Loading Time", "Epoch", "Data Loading Time")
```

```python
epoch_misc_time_list = [56.787678956985474, 59.93659019470215, 60.445101261138916, 61.81038284301758, 57.52307391166687]
plot_metrics(EPOCHS, epoch_misc_time_list, "Epoch vs Misc Time", "Epoch", "Misc Time")
```
